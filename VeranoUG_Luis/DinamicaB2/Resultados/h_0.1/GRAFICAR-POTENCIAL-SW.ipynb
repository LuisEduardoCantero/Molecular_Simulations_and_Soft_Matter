```python
from __future__ import unicode_literals
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,1])
y = np.array([10,-1.25])
x2 = np.array([1,1.25])
y2 = np.array([-1.25,-1.25])
x3 = np.array([1.25,1.25])
y3 = np.array([-1.25,0.1])
x4 = np.array([1.25,1.25+0.03])
y4 = np.array([0.1,0.1])
x5 = np.array([1.25+0.03,1.25+0.03])
y5 = np.array([0.1,0.0])

potencial100 = np.loadtxt('pot100.dat')
potencial200 = np.loadtxt('pot200.dat')
potencial300 = np.loadtxt('pot300.dat')
potencial400 = np.loadtxt('pot400.dat')
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.plot(x, y, color='k', label="SW(h=0.1)")
plt.plot(x2, y2, color='k')
plt.plot(x3, y3, color='k')
plt.plot(x4, y4, color='k')
plt.plot(x5, y5, color='k')

plt.plot(potencial100[0:,0], potencial100[0:,1], label="nk=100")
plt.plot(potencial200[0:,0], potencial200[0:,1], label="nk=200")
plt.plot(potencial300[0:,0], potencial300[0:,1], label="nk=300")
plt.plot(potencial400[0:,0], potencial400[0:,1], label="nk=400")

plt.legend()
plt.ylim(-1.3,0.2)
plt.xlim(1.2,1.3)
plt.grid(b=None, which='major', axis='both')
plt.savefig("Potenciales_0.1.jpg", bbox_inches='tight')
```

```python
epsilon = -1
H = .1

a = np.array([1,1])
b = np.array([10,epsilon])
a2 = np.array([1,1+H])
b2 = np.array([epsilon,epsilon])
a3 = np.array([1 + H,1 + H])
b3 = np.array([epsilon,0.5])
a4 = np.array([1 + H,1 + H + 0.03])
b4 = np.array([0.5,0.5])
a5 = np.array([1 + H + 0.03,1 + H + 0.03])
b5 = np.array([0.5,0.0])
a55 = np.array([1 + H + 0.03,1 + H + 0.3])
b55 = np.array([0.0,0.0])



a6 = np.array([-.1, 2])
b6 = np.array([0, 0])
a7 = np.array([0, 0])
b7 = np.array([-2, 2])
a8 = np.array([-.05, 1+H])
b8 = np.array([.5, .5])
a9 = np.array([-0.05, 1+H])
b9 = np.array([epsilon, epsilon])

```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.rcParams.update({'font.size':30})
lw=3.5
plt.plot(a, b, color='k', linewidth = lw)
plt.plot(a2, b2, color='k', linewidth = lw)
plt.plot(a3, b3, color='k', linewidth = lw)
plt.plot(a4, b4, color='k', linewidth = lw)
plt.plot(a5, b5, color='k', linewidth = lw)
plt.plot(a55, b55, color='k', linewidth = lw)
plt.plot(a6, b6, color='k', ls = '--')
plt.plot(a7, b7, color='k', ls = '--')
plt.plot(a8, b8, color='k', ls = ':')
plt.plot(a9, b9, color='k', ls = ':')

plt.text(.5, -1.15, "$\sigma$")
plt.text(-.12, 0.5-.03, "h")
plt.text(-.12, epsilon-.03, "$\epsilon$")
plt.text(1.4, -.15, "r")
plt.text(-.2, 1.0, "V(r)")
plt.text(1+H/2-.03, -1.15, "H")
plt.text(1+H/2+.12, -.32, "$\delta$")
plt.arrow(1+H+.01,0,.07,-.2)




plt.axis('off')
plt.ylim(-1.5,1)
plt.xlim(-.1,1.5)
plt.savefig("Potenciales_0.1.jpg", bbox_inches='tight')
plt.savefig("Potenciales_0.1.pdf", bbox_inches='tight')
```

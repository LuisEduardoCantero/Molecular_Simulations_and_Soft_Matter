# Computer simulation of liquids

```python
# Importamos librerias
import numpy as np
from __future__ import unicode_literals
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
```

### Potencial Yukawa

```python
rc = 1.2424
x = np.arange(1,rc,0.001)
x3 = np.arange(rc,3,1)
y3=np.array([0, 0])
z1=10
z2=0.50
lamb=0.10
tred=0.25

y1 = (1/(tred*(1-lamb)*x))*(-np.exp(-z1*(x-1))+lamb*np.exp(-z2*(x-1)))
```

```python
plt.plot(x,y1, "-", color = 'b', linewidth = 3.5)
plt.plot(x3,y3, "-", color = 'b', linewidth = 3.5)
plt.rcParams.update({'font.size':25})
plt.xlabel("r")
plt.ylabel('$\mathit{U(r)/K_{B}T}$')
plt.xlim(1, 1.4)
plt.ylim(-4.05, .5)
plt.grid(b=None, which='major', axis='both')
plt.rcParams["figure.figsize"] = (10,10)
plt.savefig("Potencial HSDYukawa.jpg", bbox_inches='tight')
```

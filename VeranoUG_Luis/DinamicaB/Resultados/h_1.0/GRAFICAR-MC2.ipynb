# Resultados de la 1a simulacion usando Dinamica Browniana, altura de la barrera (h) = 0.10

```python
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy.special
import numpy as np
import math
import pandas as pd
```

```python
potencial = np.loadtxt('pot300.dat')
g = np.loadtxt('g2.dat', skiprows=2)
e = np.loadtxt('e2.dat')
confin = np.loadtxt('conf_fin2.dat')
fself = np.loadtxt('Fself2.dat')
```

```python
bl=7
base=bl*bl/2
data = np.loadtxt('conf_fin2.dat').T
x1=np.asarray(data[0])
y1=np.asarray(data[1])
z1=np.asarray(data[2])
cx1 = x1 + base
cy1 = y1
cz1 = z1
cx2 = x1 - base
cy2 = y1
cz2 = z1
cx3 = x1
cy3 = y1 + base
cz3 = z1
cx4 = x1
cy4 = y1 - base
cz4 = z1
cx5 = x1
cy5 = y1
cz5 = z1 + base
cx6 = x1
cy6 = y1
cz6 = z1 - base
cx7 = x1 + base
cy7 = y1 + base
cz7 = z1 + base
cx8 = x1 - base
cy8 = y1 + base
cz8 = z1 + base
cx9 = x1 + base
cy9 = y1 + base
cz9 = z1 - base
cx10 = x1 - base
cy10 = y1 + base
cz10 = z1 - base
cx11 = x1 - base
cy11 = y1 - base
cz11 = z1 + base
cx12 = x1 + base
cy12 = y1 - base
cz12 = z1 + base
cx13 = x1 - base
cy13 = y1 - base
cz13 = z1 - base
cx14 = x1 + base
cy14 = y1 - base
cz14 = z1 - base

px = np.block([x1, cx1, cx2, cx3, cx4, cx5, cx6, cx7, cx8, cx9, cx10, cx11, cx12, cx13, cx14])
py = np.block([y1, cy1, cy2, cy3, cy4, cy5, cy6, cy7, cy8, cy9, cy10, cy11, cy12, cy13, cy14])
pz = np.block([z1, cz1, cz2, cz3, cz4, cz5, cz6, cz7, cz8, cz9, cz10, cz11, cz12, cz13, cz14])
```

```python
# Medicion para saber si la distancia entre particulas es menor a rc 
diametro = 0.000001
phi = 0.1
dd = math.pow(3.14159265358 * diametro * diametro * diametro / (6 * phi),1 / 3)
rc=1.25 * diametro / dd
# Definimos una funcion para calcular las distancias menores al radio rc 
aa = np.arange(len(x1))
def distancia(x1,y1,z1):
    npa = []
    i=0
    while i <= (len(x1)/15 - 1):
        vecinos = 0
        j = 0
        while j <= len(x1) - 1:
            if(j!=i):
                if(pow(((x1[i] - x1[j]) ** 2 + (y1[i] - y1[j]) ** 2 + (z1[i] - z1[j]) ** 2), 0.5) <= rc):
                    vecinos += 1                        
            j += 1
        npa.append(vecinos)
        i += 1
    return npa
```

```python
#Calculamos las distancias entre las particulas
nve = distancia(px, py, pz)

#Guardamos el numero de particula y el numero de vecinos que tiene
#c = np.array([a, nve]).T
#np.savetxt('distancias.dat', c, fmt='%i', delimiter = '\t')
```

```python
bl=7
base=bl*bl/2
data = np.loadtxt('conf_fin2.dat').T
x1=np.asarray(data[0])
y1=np.asarray(data[1])
z1=np.asarray(data[2])
cx1 = x1 + base
cy1 = y1
cz1 = z1
cx2 = x1 - base
cy2 = y1
cz2 = z1
cx3 = x1
cy3 = y1 + base
cz3 = z1
cx4 = x1
cy4 = y1 - base
cz4 = z1
cx5 = x1
cy5 = y1
cz5 = z1 + base
cx6 = x1
cy6 = y1
cz6 = z1 - base
cx7 = x1 + base
cy7 = y1 + base
cz7 = z1 + base
cx8 = x1 - base
cy8 = y1 + base
cz8 = z1 + base
cx9 = x1 + base
cy9 = y1 + base
cz9 = z1 - base
cx10 = x1 - base
cy10 = y1 + base
cz10 = z1 - base
cx11 = x1 - base
cy11 = y1 - base
cz11 = z1 + base
cx12 = x1 + base
cy12 = y1 - base
cz12 = z1 + base
cx13 = x1 - base
cy13 = y1 - base
cz13 = z1 - base
cx14 = x1 + base
cy14 = y1 - base
cz14 = z1 - base
cx15 = x1 
cy15 = y1 + base
cz15 = z1 + base
cx16 = x1 + base
cy16 = y1 + base
cz16 = z1 
cx17 = x1 - base
cy17 = y1 + base
cz17 = z1 
cx18 = x1 
cy18 = y1 + base
cz18 = z1 - base
cx19 = x1 
cy19 = y1 - base
cz19 = z1 - base
cx20 = x1 - base
cy20 = y1 - base
cz20 = z1 
cx21 = x1 + base
cy21 = y1 - base
cz21 = z1 
cx22 = x1 
cy22 = y1 - base
cz22 = z1 + base
cx23 = x1 + base
cy23 = y1 
cz23 = z1 + base
cx24 = x1 - base
cy24 = y1 
cz24 = z1 + base
cx25 = x1 + base
cy25 = y1 
cz25 = z1 - base
cx26 = x1 - base
cy26 = y1 
cz26 = z1 - base

px = np.block([x1, cx1, cx2, cx3, cx4, cx5, cx6, cx7, cx8, cx9, cx10, cx11, cx12, cx13, cx14, cx15, cx16, cx17, cx18, cx19, cx20, cx21, cx22, cx23, cx24, cx25, cx26])
py = np.block([y1, cy1, cy2, cy3, cy4, cy5, cy6, cy7, cy8, cy9, cy10, cy11, cy12, cy13, cy14, cy15, cy16, cy17, cy18, cy19, cy20, cy21, cy22, cy23, cy24, cy25, cy26])
pz = np.block([z1, cz1, cz2, cz3, cz4, cz5, cz6, cz7, cz8, cz9, cz10, cz11, cz12, cz13, cz14, cz15, cz16, cz17, cz18, cz19, cz20, cz21, cz22, cz23, cz24, cz25, cz26])
```

```python
len(x1)
```

```python
# Medicion para saber si la distancia entre particulas es menor a rc 
diametro = 0.000001
phi = 0.1
dd = math.pow(3.14159265358 * diametro * diametro * diametro / (6 * phi),1 / 3)
rc=1.25 * diametro / dd
# Definimos una funcion para calcular las distancias menores al radio rc 
a = np.arange(len(x1))
def distancia(x1,y1,z1):
    npa = []
    vx = []
    vy = []
    vz = []
    d = []
    i=0
    while i <= (len(x1)/27 - 1):
        vecinos = 0
        j = 0
        while j <= len(x1) - 1:
            if(j!=i):
                d.append(pow(((x1[i] - x1[j]) ** 2 + (y1[i] - y1[j]) ** 2 + (z1[i] - z1[j]) ** 2), 0.5))
                if(pow(((x1[i] - x1[j]) ** 2 + (y1[i] - y1[j]) ** 2 + (z1[i] - z1[j]) ** 2), 0.5) <= rc):
                    vecinos += 1
                    if(j>=len(x1)/27):
                        vx.append(x1[j])
                        vy.append(y1[j])
                        vz.append(z1[j])
            j += 1
        npa.append(vecinos)
        i += 1
    return npa, vx, vy, vz, d
```

```python
1.25 * diametro / dd
```

```python
#Calculamos las distancias entre las particulas
nve, vx, vy, vz, d = distancia(px, py, pz)

#Guardamos el numero de particula y el numero de vecinos que tiene
#c = np.array([a, nve]).T
#np.savetxt('distancias.dat', c, fmt='%i', delimiter = '\t')
```

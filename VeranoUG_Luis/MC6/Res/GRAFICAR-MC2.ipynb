```python
import numpy as np
from matplotlib import pyplot
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
```

```python
potencial = np.loadtxt('pot.dat')
g = np.loadtxt('g.dat')
e = np.loadtxt('e.dat')
confin = np.loadtxt('conf_fin.dat')
```

```python
plt.plot(potencial[450:,0], potencial[450:,1])
plt.ylim(-1.9,0.3)
plt.xlim(0.75,1.4)
```

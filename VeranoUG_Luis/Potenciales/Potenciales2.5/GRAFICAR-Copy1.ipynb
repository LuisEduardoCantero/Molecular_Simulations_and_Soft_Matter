```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,1])
y = np.array([10,-1.25])
x2 = np.array([1,1.25])
y2 = np.array([-1.25,-1.25])
x3 = np.array([1.25,1.25])
y3 = np.array([-1.25,2.5])
x4 = np.array([1.25,1.25+0.03])
y4 = np.array([2.5,2.5])
x5 = np.array([1.25+0.03,1.25+0.03])
y5 = np.array([2.5,0.0])

potencial100 = np.loadtxt('pot100.dat')
potencial200 = np.loadtxt('pot200.dat')
potencial300 = np.loadtxt('pot300.dat')
potencial400 = np.loadtxt('pot400.dat')
```

```python
plt.rcParams["figure.figsize"] = (10,10)
plt.plot(x, y, color='k', label="SW(h=2.5)")
plt.plot(x2, y2, color='k')
plt.plot(x3, y3, color='k')
plt.plot(x4, y4, color='k')
plt.plot(x5, y5, color='k')

plt.plot(potencial100[0:,0], potencial100[0:,1], label="nk=100")
plt.plot(potencial200[0:,0], potencial200[0:,1], label="nk=200")
plt.plot(potencial300[0:,0], potencial300[0:,1], label="nk=300")
plt.plot(potencial400[0:,0], potencial400[0:,1], label="nk=400")

plt.legend()
plt.ylim(-1.75,4.2)
plt.xlim(1.2,1.3)
plt.grid(b=None, which='major', axis='both')
plt.savefig("Potenciales_2.5.jpg", bbox_inches='tight')
```
